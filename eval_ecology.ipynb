```python
import json
import csv
import time
import random
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diss_dir = "/Volumes/USB_W/DISS/DOI_detector/ecology/BettysResult_ecology.json"
```

```python
file = open(diss_dir)
search_results = json.load(file)
file.close()
```

```python
search_results[0]["name"]
```

```python
def clear_output():
    from IPython.display import clear_output
    clear_output(wait=False)
```

```python
list_of_idx = [i for i in range(len(search_results))]
random.shuffle(list_of_idx)
print(list_of_idx)
```

```python
# save list in file
with open("/Volumes/USB_W/DISS/DOI_detector/ecology" + "/list_of_idx.txt", "w") as file:
    for item in list_of_idx:
        file.write("%s\n" % item)
    print("Done")
```

# first 100 repos

```python
repos_with_reference = []
eval_results = []
kind_of_reference = []

num_of_repos = 400
for i, idx in enumerate(list_of_idx[0:100]):
    #clear_output()
    #time.sleep(1)

    print(" +++++++++ ")
    print(" ")
    
    print(f"Progress :  {num_of_repos - i}")
    print(f"Das hier ist Repository {idx}")
    print("Name des Repos : " + search_results[idx]["name"])
    print("Link zum Repository " + search_results[idx]["url"])
    try:
        print("Titel des Main Papiers : " + search_results[idx]["mainPaper"]["title"])
    except:
        print("Titel des Main Papiers : " + "Kein Main Paper gefunden")

    print("zenodo : " + search_results[idx]["agentQueryTerm"]["zenodo"])
    print("openAlex : " + search_results[idx]["agentQueryTerm"]["openAlex"])
    print("openCitations : " + search_results[idx]["agentQueryTerm"]["openCitations"])
    print("semanticScholar : " + search_results[idx]["agentQueryTerm"]["semanticScholar"])
    print("dataCite : " + search_results[idx]["agentQueryTerm"]["dataCite"])
    print("Number of Publicaitons : " + str(len(search_results[idx]["publications"])))


    print(f"Bisherige Arten von Referenzen : {set(kind_of_reference)}")
        
    print(" ")
    print(" -------- ")
    print(" ")

    kind_of_reference_ = input("Wie wurde referenziert? (None fr garnicht)")

    repo_has_reference = input("Verweist das Repo auf eine eigene Arbeit (t/f/a) ? : ")

    print("das repo hat einen verweis & wir haben eine kask suche durchgefhrt : tp")
    print("das repo hat keinen verweis & wir haben keine kask suche durchgefhrt : tn")
    print("das repo hat keinen verweis & wir haben eine kask suche durchgefhrt : fp")
    print("das repo hat einen verweis & wir haben keine kask suche durchgefhrt : fn")

    kask_search = input("tp/tn/fp/fn : ")
    
    
    if repo_has_reference == 'a':
        break
        
    repos_with_reference.append(repo_has_reference)
    eval_results.append(kask_search)
    kind_of_reference.append(kind_of_reference_)
    
```

```python
kask_search_tp = eval_results.count("tp")
kask_search_tn = eval_results.count("tn")
kask_search_fp = eval_results.count("fp")
kask_search_fn = eval_results.count("fn")

print("das repo hat einen verweis & wir haben eine kask suche durchgefhrt : " + str(kask_search_tp))
print("das repo hat keinen verweis & wir haben keine kask suche durchgefhrt : " + str(kask_search_tn))
print("das repo hat keinen verweis & wir haben eine kask suche durchgefhrt : " + str(kask_search_fp))
print("das repo hat einen verweis & wir haben keine kask suche durchgefhrt : " + str(kask_search_fn))

print(f"{len(repos_with_reference)} Repos have a reference")

eval_set_of_references = []
set_of_references = set(kind_of_reference)
for element in set_of_references:
    count_references = kind_of_reference.count(element)
    eval_set_of_references.append([element, count_references])


print(eval_set_of_references)


```

```python
array = [[kask_search_tp,kask_search_fp],[kask_search_fn,kask_search_tn]]
df_cm = pd.DataFrame(array, index=["True", "False"], columns=["True", "False"])
plt.figure(figsize = (10,7))
sn.heatmap(df_cm, annot=True, cmap="crest", linewidth=.5)

print(eval_set_of_references)
eval_set_sorted = sorted(eval_set_of_references, key=lambda x: x[1])[::-1]


```

```python
x = []
y = []
for i in eval_set_sorted:
    x.append(i[0])
    y.append(i[1])

plt.bar(x, y) 

a = repos_with_reference.count("t")
b = repos_with_reference.count("f")
print(f"repos with references : {a}")
print(f"repos without references : {b}")
```
